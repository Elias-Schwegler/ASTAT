### <span style="color:rgb(160,0,86)">Wie können wir die Verteilung von Daten graphisch darstellen?</span>

Lagemasse und Streumasse fassen die Daten mit einem typischen Wert zusammen. Aber es ist auch wertvoll, die **gesamte Verteilung** der Daten zu untersuchen. 
- Die Verteilung ***nominaler Merkmale*** wird häufig in einem <span style="color:rgb(160,0,86)">**Kreisdiagramm**</span> dargestellt. Die relativen Häufigkeiten aller Werte des Merkmals werden proportional als Kreissektoren abgebiltet.  
- Für ***ordinale Markmale*** ist ein <span style="color:rgb(160,0,86)">**Balkendiagramm**</span> besser geeignet, weil die natürliche Reihenfole der Werte von links nach rechts dargestellt werden kann. Die absoluten oder relativen Häufigkeiten der Werte werden proportional zur Höhe eines Balken abgebildet.
- Die Verteilung ***metrischer Merkmale*** werden in <span style="color:rgb(160,0,86)">**Histogrammen**</span> oder <span style="color:rgb(160,0,86)">**Dichtediagrammen**</span> dargestellt, in denen die vielen Zahlenwerte des Merkmals in Klassen eingeteilt werden. Eine weitere häufige Visualisierung von metrischen Daten ist der <span style="color:rgb(160,0,86)">**Box-Plot**</span>, in dem neben der *Verteilung* auch ein *Lagemass* (Median) und ein *Streumass* (Quartielsabstand) sichtbar werden.

#### <span style="color:rgb(160,0,86)">Beispiel:</span>

Wir betrachten wieder die 10 ehemaligen Studierenden:

$$\begin{bmatrix}
\text{Anna}&w&\text{ausgezeichnet}&27\\
\text{Beat}&m&\text{gut}&34\\
\text{Cary}&m&\text{sehr gut}&29\\
\text{Dana}&w&\text{sehr gut}&24\\
\text{Elif}&w&\text{gut}&25\\
\text{Faro}&m&\text{ausgezeichnet}&27\\
\text{Gabi}&w&\text{sehr gut}&27\\
\text{Hans}&m&\text{genügend}&69\\
\text{Ivea}&w&\text{sehr gut}&26\\
\text{Jose}&w&\text{gut}&31
\end{bmatrix}$$ 

Wir erstellen wieder mit **pandas** ein DataFrame Objekt mit Namen **df** für diese Daten:

```python
namen = ["Anna","Beat","Cary","Dana","Elif","Faro","Gabi","Hans","Ivea","Jose"]
geschlecht = ["w","m","m","w","w","m","w","m","w","w"]
prädikat = ["ausgezeichnet","gut","sehr gut","sehr gut","gut","ausgezeichnet","sehr gut","genügend","sehr gut","gut"]
alter = [27,34,29,24,25,27,27,69,26,31]

import numpy as np
import pandas as pd
df = pd.DataFrame({"Name":namen,"Geschlecht":geschlecht,
                   "Prädikat":prädikat,"Alter":alter})
```

#### <span style="color:rgb(160,0,86)">Wie können wir mit Python Diagramme Zeichnen?</span>

Das Modul **Matplotlib** ist das wichtigste **Visualisierungsmodul** in Python, es wird oft als ***Standardbibliothek für Plots*** bezeichnet.

- Es erlaubt die Erstellung von Diagrammen wie Linien-, Balken-, Punktdiagramme, Histogramme und viele mehr.
- Link zur **Modulseite**: https://matplotlib.org/

Der Teilbereich **matplotlib.pyplot** stellt eine einfache, an MATLAB angelehnte Schnittstelle bereit:

- Jeder Befehl verändert eine aktuelle Figur (engl. ***figure***) oder eine Achse (engl. ***axis***).
- Link zur **API-Reference**: https://matplotlib.org/stable/api/pyplot_summary.html

#### <span style="color:rgb(160,0,86)">Kreisdiagramme</span>

Mit der Funktion **pie()** aus dem Teilbereich **matplotlib.pyplot** können wir die Häufigkeiten des **nominalen** Merkamls **Geschlecht** sehr einfache als **Kreisdiagramm** darstellen:

```python
werte,absH = np.unique(df["Geschlecht"],return_counts=True)

#pip install matplotlib
import matplotlib.pyplot as plt

plt.pie(absH,labels=werte,startangle=90,autopct="%1.0f%%")
plt.show()
```

***Bemerkung:*** Der Parameter **autopct=**<tt>FORMAT</tt> in der Funktion **plt.pie()** steuert, wie die Prozentwerte in den Kreissegmenten angezeigt werden. In der Zeichenkette (string) <tt>FORMAT</tt> wird das Ausgabeformat codiert:

Ein **Format-String** in Python hat die Form: 
$$\texttt{\%}\,[\texttt{Breite}].[\texttt{Nachkommastellen}]\,\texttt{Zahlausgabe}$$
- $\texttt{\%}$ → Platzhalter für den Wert
- $\texttt{Breite}\;=\;i$ → String ist mindestens $\,i\,$ Zeichen breit
- $\texttt{Nachkommastellen}\;=\;i$ Wert hat $\,i\,$ Nachkkommastellen
- $\texttt{Zahlausgabe}\;=\;$ **f** → Zahl im Fliesskomma-Format (float) <br>
  $\texttt{Zahlausgabe}\;=\;$ **d** → Zahl als ganze Zahl (integer) <br>
  $\texttt{Zahlausgabe}\;=\;$ **e** → Zahl im Exponentialformat
- $\texttt{\%\%}$ → Das doppelte Prozentzeichen sorgt dafür, dass wirklich ein **%** Zeichen ausgegeben wird (sonst würde Python es als Platzhalter interpretieren).

```python
geschlecht_Disp0 = ["w","w","w","w","w","w","w","w","w","w"]
geschlecht_Disp1 = ["w","m","m","w","m","m","w","m","w","w"]

df_Disp0 = pd.DataFrame({"Name":namen,"Geschlecht":geschlecht_Disp0})
df_Disp1 = pd.DataFrame({"Name":namen,"Geschlecht":geschlecht_Disp1})

werte_Disp0,absH_Disp0 = np.unique(df_Disp0["Geschlecht"],return_counts=True)
werte_Disp1,absH_Disp1 = np.unique(df_Disp1["Geschlecht"],return_counts=True)

fig, ax = plt.subplots(1, 2, figsize=(10, 5))  # 1 Zeile, 2 Spalten

# Erstes Kuchendiagramm
ax[0].pie(absH_Disp0,
          labels=werte_Disp0,
          startangle=90,
          autopct="%1.0f%%")
ax[0].set_title(r"Dispersionsindex $P = 0$ (keine Streuung)")

# Zweites Kuchendiagramm
ax[1].pie(absH_Disp1,
          labels=werte_Disp1,
          startangle=90,
          autopct="%1.0f%%")
ax[1].set_title(r"Dispersionsindex $P = 1$ (maximale Streuung)")

plt.show()
```

#### <span style="color:rgb(160,0,86)">Balkendagramme</span>
Die Verteilung des **ordinalen** Merkamls **Prädikat** können wir mit einem **Balkendiagramm** darstellen:

```python
# pip install seaborn
import seaborn as sns
import matplotlib.pyplot as plt

fig = plt.figure()

ax = fig.add_subplot(1,1,1) 
# 1.Zahl -> Zeile, 2. Zahl -> Spalte, 3. Zahl -> Position
sns.countplot(data=df,x="Prädikat",
              order=["genügend","gut","sehr gut","ausgezeichnet"],ax = ax)
ax.set_title("Balkendiagramm")
ax.set_ylabel("Absolute Häufigkeit")
ax.set_xlabel("Prädikat der Studierenden FS25")

plt.show()
```

***Bemerkung:*** **Wie weit** die Werte des Markmals **Prädikat** auseinander liegen, ist **nicht relevant**. Sie sind nur von <tt>'genügend'</tt> bis <tt>'ausgezeichnet'</tt> geordnet.

```python
prädikat_Div0 = ["gut","gut","gut","gut","gut","gut","gut","gut","gut","gut"]
prädikat_Div1 = ["ausgezeichnet","ausgezeichnet","ausgezeichnet","genügend","genügend","ausgezeichnet","genügend","genügend","ausgezeichnet","genügend"]

df_Div0 = pd.DataFrame({"Name":namen,"Prädikat":prädikat_Div0})
df_Div1 = pd.DataFrame({"Name":namen,"Prädikat":prädikat_Div1})

fig, ax = plt.subplots(1, 2, figsize=(10, 5))  # 1 Zeile, 2 Spalten

# Erstes Balkendiagramm
sns.countplot(data=df_Div0,x="Prädikat",
              order=["genügend","gut","sehr gut","ausgezeichnet"],ax = ax[0])
ax[0].set_title(r"Diversität $D = 0$ (keine Streuung)")
ax[0].set_ylabel("Absolute Häufigkeit")
ax[0].set_xlabel("Prädikat der Studierenden")

# Zweites Balkendiagramm
sns.countplot(data=df_Div1,x="Prädikat",
              order=["genügend","gut","sehr gut","ausgezeichnet"],ax = ax[1])
ax[1].set_title("Diversität $D = 1$ (maximale Streuung)")
ax[1].set_ylabel("Absolute Häufigkeit")
ax[1].set_ylim(0,10)
ax[1].set_xlabel("Prädikat der Studierenden")

plt.show()
```

#### <span style="color:rgb(160,0,86)">Histogramme und Dichtediagramme</span>

Die ***Abstände zwischen Werten*** eines **metrischen** Merkmales sind im Unterschied zu ordinalen Merkmalen immer eindeutig gegeben. 

Oft haben metrische Merkmale auch ***viele verschiedene Werte***, so dass ein Balkendiagramm die Charakteristik einer Verteilung nicht mehr sinnvoll abbildet. 

Daher ***unterteilen wir die Werte*** eines metrischen Merkmals in Abschitte, sogenannte ***Klassen***, und bilden die absoluten oder relativen Häufigkeiten der Werte in diesen Klassen. Eine gleichmässige Klasseneinteilung für das Merkmal **Alter** mit einer **Klassenbreite** von 5 Jahren ist zum Beispiel:

$$\begin{array}{l|cc}
\text{Klassen}&n(\;) & h(\;)\\\hline
\text{von 20 bis 24} & 1 & 0.1 \\
\text{von 25 bis 29} & 6 & 0.6 \\
\text{von 30 bis 34} & 2 & 0.2 \\
\text{von 35 bis 39} & 0 & 0.0 \\
\text{von 40 bis 44} & 0 & 0.0 \\
\text{von 45 bis 49} & 0 & 0.0 \\
\text{von 50 bis 54} & 0 & 0.0 \\
\text{von 55 bis 59} & 0 & 0.0 \\
\text{von 60 bis 64} & 0 & 0.0 \\
\text{von 65 bis 69} & 1 & 0.1 \\
\end{array}$$

Wenn also Häufigkeiten über einer Unterteilung der Zahlenachse abgebildet werden, entsteht ein **Histogramm** oder ein **Dichtediagramm**:

- In **Histogrammen** werden die Häufigkeiten der Klassen ***proportional zur Höhe der Rechtecke*** abgebildet.

***Bemerkung:*** Die **Wahl der Klasseneinteilung** ist sehr entscheidend, damit die Struktur der Häufigkeitsverteilung sichtbar wird! Mit dem Parameter **bins** können wir die Einteilung angeben:

- $\texttt{bins}\;=\;i$ → es werden $\,i\,$ gleich breite Klassen gebildet
- $\texttt{bins}\;=\;\texttt{list}$ → die Klassengrezen werden als Liste angegeben

Ohne Angabe des Parameters **bins** wird standardmässig ein Algorithmus für die automatische Klasseneinteilung angewendet:

- $\texttt{bins}\;=\;\texttt{auto}$ → NumPy wählt zwischen der **Sturges-Regel** (für kleinere Datensätze) und der **Freedman–Diaconis-Regel** (für grössere oder stark variierende Datensätze). <br>
  So wird versucht eine sinnvolle Balance zwischen *zu vielen* und *zu wenigen* Klassen zu finden.

```python
import seaborn as sns
import matplotlib.pyplot as plt

fig = plt.figure(figsize=(12,12))

# 1. Zahl -> Zeile, 2. Zahl -> Spalte, 3. Zahl -> Position
ax1 = fig.add_subplot(2,2,1) 
sns.histplot(data=df,x="Alter",bins=2,ax=ax1)
ax1.set_title("Histogramm mit 2 Klassen")
ax1.set_ylabel("Absolute Häufigkeit")

ax2 = fig.add_subplot(2,2,2)
sns.histplot(data=df,x="Alter",bins=3,ax=ax2)
ax2.set_title("Histogramm mit 3 Klassen")
ax2.set_ylabel("Absolute Häufigkeit")

ax3 = fig.add_subplot(2,2,3)
sns.histplot(data=df,x="Alter",bins=5,ax=ax3)
ax3.set_title("Histogramm mit 5 Klassen")
ax3.set_ylabel("Absolute Häufigkeit")

ax4 = fig.add_subplot(2,2,4)
sns.histplot(data=df,x="Alter",ax=ax4)
ax4.set_title("Automatische Klasseneinteilung")
ax4.set_ylabel("Absolute Häufigkeit")

plt.show()
```

- In **Dichtediagrammen** werden die relativen Häufgikeiten der Klassen ***proportional zur Fläche der Rechtecke*** abgebildet. Das können wir mit dem Parameter **stat=**<tt>"density"</tt> einstellen: 

```python
import seaborn as sns
import matplotlib.pyplot as plt

fig = plt.figure(figsize=(12,5))

ax1 = fig.add_subplot(1,2,1)
sns.histplot(data=df,x="Alter",bins=2,stat="density",color="red",ax=ax1)
ax1.set_title("Dichtediagramm mit 2 Klassen")
ax1.set_ylabel("Dichte")

ax2 = fig.add_subplot(1,2,2)
sns.histplot(data=df,x="Alter",bins=np.arange(23,70,1),stat="density",color="red",ax=ax2)
ax2.set_title("Dichtediagramm mit Klassen der Breite 1 Jahr")
ax2.set_ylabel("Dichte")

plt.show()
```

***Bemerkung:*** 

<span style="color:rgb(160,0,86)">**Die Gesamtfläche aller Rechtecke in einem Dichtediagramm beträgt immer 1 bzw. 100%.**</span>

#### <span style="color:rgb(160,0,86)">Typische Formen von Verteilungen</span>

Ein Histogramm oder Dichtediagramm heisst:
- **symmetrisch**, wenn die meisten Daten um den Mittelwert liegen und ungefähr die hälfte der Daten unter beziehungsweise über dieser mittleren Lage sind.
- **linksschief**, wenn die Daten weiter nach links abfallen. Es gibt also auf der linken Seiten extreme Werte.
- **rechtsschief**, wenn die Daten weiter nach rechts abfallen. Es gibt also auf der rechten Seiten extreme Werte.

Das Merkmal **Alter** hat demnach eine ***rechtsschiefe*** Verteilung.

#### <span style="color:rgb(160,0,86)">Box-Plots</span>

Um die **Lage der Daten** und die **Symmetrie** eines metrischen Merkmals graphsich darzustellen, können wir einen sogenannten **Box-Plot** zeichen:
- mit einem Rechteck, dessen Höhe dem Quartilsabstand $Q$ enspricht
- der Median $x_{50}$ unterteil das Rechteck in zwei Teile
- unten und oben wird mit einem ***Whisker*** der kleinest beziehungsweise der grösste noch normale Wert angegeben.
- Werte, die $1.5\,Q$-mal weiter vom unteren beziehungsweise vom oberen Quartil enfernt sind, heissen ***Ausreisser*** und sind nicht mehr innerhalb der beiden Whisker. Solche extreme Werte werden im Box-Plot mit einzelnen Punkten abgebildet.   

```python
import seaborn as sns
import matplotlib.pyplot as plt

fig = plt.figure(figsize=(5,8))

ax = fig.add_subplot(1,1,1)
# Über dieser Linie sind Werte Ausreisser nach oben
ax.axhline(30.5+1.5*(30.5-26), color="red",alpha=0.3)
# Unter dieser Linie sind Wert Ausreisser nach unten
ax.axhline(26-1.5*(30.5-26), color="red",alpha=0.3)
# Das obere Quartil (75%-Quantil)
ax.axhline(30.5, color="red",alpha=0.3, dashes=(2,2))
# Der Median (50%-Quantil)
ax.axhline(27, color="blue",alpha=0.3, dashes=(2,2))
# Das untere Quartil (25%-Quantil)
ax.axhline(26, color="red",alpha = 0.3, dashes=(2,2))
# Box-Plot erstellen
sns.boxplot(data=df,y="Alter",
            width=0.25,
            color="lightblue",
            flierprops={"marker":"x"},
            ax=ax)
# Zusätzlich die Daten als Punkte in einem angeordneten Schwarm
sns.swarmplot(data=df,y="Alter",color="black",ax=ax) 
# Zusätzlich die Daten als Punkte horizontal zufällig angeordnet
#sns.stripplot(data=df,y="Alter",color="black",ax = ax) 
ax.set_title("Box-Plot")

plt.show()
```

Wenn wir den **Box-Plot horizontal** zeichnen wollen, müssen wir die Daten des Merkamls **Alter** dem Parameter **x** statt dem Parameter **y** übergeben:

```python
import seaborn as sns
import matplotlib.pyplot as plt

fig = plt.figure(figsize=(8,5))

ax = fig.add_subplot(1,1,1)
sns.boxplot(data=df,x="Alter",
            width=0.25,
            color="lightblue",
            flierprops={"marker":"x"},
            ax=ax)
# Zusätzlich die Daten als Punkte horizontal zufällig angeordnet
sns.stripplot(data=df,x="Alter",color="black",ax = ax) 
ax.set_title("Box-Plot")

plt.show()
```

#### <span style="color:rgb(160,0,86)">Regenwolken-Plot:</span>

Wenn in einem **Box-Plot** die Daten ***zusätzlich als Punkte*** dargestellt werden, entsteht ein sogenannter **Regenwolken-Plot**.

Über diese Punkte können zudem auch mit ***unterschiedlichen Farben*** und ***Symbolen*** **weitere Merkamle** abgebildet werden. 


#### <span style="color:rgb(160,0,86)">Beispiel: Palmer Penguins</span>

Das **Palmer Penguins** Dataset ist ein offener Datensatz aus der Antarktis. Die Daten stammen von ökologischen Feldstudien auf den ***Palmer-Stationen*** in der Antarktis und enthalten Messungen an **drei Pinguinarten**: <tt>'Adelie'</tt>, <tt>'Chinstrap'</tt> und <tt>'Gentoo'</tt>. 

Erfasst wurden **biologische Merkmale** wie:
- **Körpergewicht**
- **Flossenlänge**
- **Schnabellänge**
- **Schnabeltiefe**
- **Geschlecht**
- **Insel** (Lebensraum)

Der Palmer Penguins Datensatz ist direkt in Seaborn enthalten und kann mit der Funktion **sns.load_dataset()** aus dem **seaborn** Modul geladen werden:

```python
import seaborn as sns
import matplotlib.pyplot as plt

penguins = sns.load_dataset("penguins")
penguins.head()
```

Wir löschen mit der Methode **dropna()** Spalten und Zeilen mit unbekannten Werten:

```python
penguins.dropna(inplace=True)
penguins.head()
```

Zu herauszufinden, welche **(species, island) Kombinationen** es im Datensatz gibt, können wir mit der Methode **drop_duplicates()** im Datesatz in diesen Spalten die Duplikate löschen: 

```python
penguins[["species","island"]].drop_duplicates()
```

Für diese ***fünf Kombinationen*** erstellen wir nun ***eigene DataFrames*** mit je den Merkmalen **Schnabellänge**, **Schnabeltiefe**, **Flossenlänge**, **Körpergewicht** und **Geschlecht**:

```python
Merkmale = ["bill_length_mm","bill_depth_mm",
            "flipper_length_mm","body_mass_g","sex"]
AdelieBiscoe = penguins[Merkmale][(penguins["species"]=="Adelie") &
                                  (penguins["island"]=="Biscoe")]
AdelieDream = penguins[Merkmale][(penguins["species"]=="Adelie") &
                                  (penguins["island"]=="Dream")]
AdelieTorgersen = penguins[Merkmale][(penguins["species"]=="Adelie") &
                                  (penguins["island"]=="Torgersen")]
ChinstrapDream = penguins[Merkmale][(penguins["species"]=="Chinstrap") &
                                     (penguins["island"]=="Dream")]
GentooBiscoe = penguins[Merkmale][(penguins["species"]=="Gentoo") &
                                  (penguins["island"]=="Biscoe")]
```

Nun erstellen wir **Box-Plots** für die Merkmale **Schnabellänge**, **Schnabeltiefe**, **Flossenlänge** und **Körpergewicht**. Für all diese Merkmale zeichen wir je einen Box-Plot für die Werte <tt>'male'</tt> und <tt>'female'</tt> und brauchen die folgende **Farb-** und **Symbolcodierung**:
- Daten von der Art ***Adelie*** werden <span style="color: red;">***rot***</span> dargestellt
- Daten von der Art ***Chinstrap*** werden <span style="color: green;">***grün***</span> dargestellt
- Daten von der Art ***Gentoo*** werden <span style="color: blue;">***blau***</span> dargestellt
- Daten von der Insel ***Biscoe*** werden mit ***Kreisen*** $\Large\circ$ dargestellt
- Daten von der Insel ***Dream*** werden mit ***Dreiecken*** $\small\triangle$ dargestellt
- Daten von der Insel ***Torgersen*** werden mit ***Sternen*** $\large\ast$ dargestellt

```python
def plot_teildaten(axi,merkmal,title, 
                   P  = penguins,
                   AB = AdelieBiscoe,
                   AD = AdelieDream, 
                   AT = AdelieTorgersen,
                   CD = ChinstrapDream, 
                   GB = GentooBiscoe):
    # Boxplot
    sns.boxplot(x="sex",y=merkmal,data=P,
                width=0.7,
                flierprops={"marker": "x"},
                color=".9", 
                ax=axi)
    axi.tick_params(axis="x",labelsize=18)

    # Stripplots (Gruppe,Farbe,Marker,Grösse)
    teildaten_codierung = [(AB,"red",   "o", 4),
                           (AD,"red",   "^", 5),
                           (AT,"red",   "*", 5),
                           (CD,"green", "^", 5),
                           (GB,"blue",  "o", 4)]

    for df,farbe,symbol,groesse in teildaten_codierung:
        sns.stripplot(x="sex",y=merkmal,order=["Male","Female"],
                      data=df[["sex",merkmal]],
                      color=farbe, 
                      marker=symbol,
                      alpha=0.5, 
                      size=groesse, 
                      ax=axi)

    axi.set_title(title,fontsize=20,pad=10)
    axi.set_xlabel("")
    axi.set_ylabel("")

# Anwendung auf 4 Subplots
fig, axes = plt.subplots(2, 2, figsize=(15, 15))

plot_teildaten(axes[0,0], "bill_length_mm", "Schnabellänge")
plot_teildaten(axes[0,1], "bill_depth_mm", "Schnabeltiefe")
plot_teildaten(axes[1,0], "flipper_length_mm", "Flügellänge")
plot_teildaten(axes[1,1], "body_mass_g", "Körpergewicht")

plt.tight_layout()
plt.show()
```

### <span style="color:rgb(160,0,86)">Aufgabe 3</span>

Laden Sie die Daten *2022_Mathematik 1 WiSo Urliste* und zeichnen Sie für die Punkteverteilung der Aufgaben passende Diagramme. 

Bestimmen Sie gemäss der Formel $$\frac{\text{Erreichte Anzahl Punkte}}{\text{maximale Anzahl Punkte}}\cdot 5 + 1$$ für alle Studierende die Note und stellen Sie die Verteilung der Noten graphisch dar. Bestimmen Sie auch passende Lage- und Streumasse für die Noten.

***

### <span style="color:rgb(160,0,86)">Lösungsvorschlag</span>

Die Daten können wir genau gleich laden wie in der Serie 02:


```python
# pip install xlrd
punkte = pd.read_excel("Daten/2022_Mathematik 1 WiSo Urliste.xls")

# wir speichern die bestehenden Namen der Spalten
colNamen = punkte.columns

# und schneiden die relevanten Spalten (axis=1) und Zeilen (axis=0) heraus
punkte.drop(list(colNamen[0:1])+list(colNamen[10:11])+list(colNamen[12:]),axis=1,inplace=True)
punkte.drop([0,1,2,3],axis=0,inplace=True)

# nun geben wir den Spalten einfache Namen
colNamenNew = ["A1","A2","A3","A4","A5","A6","A7","A8","A9","Version"]
punkte.columns = colNamenNew

# dann speichern wir die Studierenden mit Veresion A
VersionA = punkte[colNamenNew][punkte["Version"]=="A"]

# bei den Studierenden mit Version B müssen wir die Reihenfolge der Aufgaben ändern
VersionB = punkte[["A2","A1","A4",
                   "A3","A6","A5",
                   "A8","A7","A9","Version"]][punkte["Version"]=="B"]
VersionB.columns = colNamenNew

# zuletzt fügen wir die zwei DataFrame VersionA und VersionB wieder zusammen
punkteNew = pd.concat([VersionA,VersionB])

punkteNew
```

Neu fügen wir nun die Spalten **Summe**, **Note** und **Note gerundet** hinzu:

```python
# Wir fügen die Spalte "Summe" für die Punktesumme hinzu:
punkteNew["Summe"] = punkteNew[["A1","A2","A3","A4","A5","A6",
                                "A7","A8","A9"]].sum(axis=1)

# Wir fügen die Spalte "Note" mit der ungerundeten Note hinzu:
punkteMax = 36
punkteNew["Note"] = punkteNew["Summe"]/punkteMax*5+1

# Die Funktion rundenSym() können wir die Noten auf halbe Noten runden:
def rundenSym(note):
    return np.floor(2*note+0.5)/2
    
# Wir fügen die Spalte "Note gerundet" hinzu:
punkteNew["Note gerundet"] = punkteNew["Note"].apply(rundenSym)

punkteNew
```

- Um die Verteilung der Punkte bei jeder Aufgabe zu visualisieren, eignen sicht **Boxplots**, damit wir die Punkteverteilung von allen 9 Aufgaben in einer Übersicht erfassen und vergleichen können.
- Damit wir für alle 9 Aufgaben in der gleichen Figur einen Box-Plot zeichnen können, wandeln wir mit der Funktion **pandas.melt()** wie bei der Aufgabe 2 das DataFrames **punkteNew** vom ***wide*** ins ***long format***:

```python
punkte_long = punkteNew[["A1","A2","A3","A4","A5","A6","A7","A8","A9"]].melt(var_name="Bereich", value_name="Punkte")

punkte_long
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

fig = plt.figure(figsize=(12,5))

ax = fig.add_subplot(1,1,1)

sns.boxplot(data=punkte_long,
            x="Bereich",
            y="Punkte",
            width=0.7,
            flierprops={"marker":"x"},
            ax=ax)

# Farben ändern
farben = sns.light_palette("seagreen", n_colors=len(ax.patches))
for patch, farbe in zip(ax.patches, farben):
    patch.set_facecolor(farbe)
    
ax.set_title("Mathematikprüfung I an der WISO-Fakultät der Uni Bern",fontsize=14,pad=15)
ax.set_xlabel("Aufgaben",fontsize=12,labelpad=15)
ax.set_ylabel("Punkte",fontsize=12,labelpad=15)

plt.show()
```

- Für die Visualisierung der Notenverteilung ist zusätzlich ein Histogramm geeignet, weil wir diese Gesamtnoten nicht mit anderen Daten vergleichen.

```python
fig = plt.figure(figsize=(13,5))

mittelwert = punkteNew["Note gerundet"].mean()
std = punkteNew["Note gerundet"].std(ddof=0)
x_50 = punkteNew["Note gerundet"].quantile(0.5)
IQR = (punkteNew["Note gerundet"].quantile(0.75) - 
       punkteNew["Note gerundet"].quantile(0.25))

ax1 = fig.add_subplot(1,2,1)
sns.histplot(data=punkteNew,
             x="Note gerundet",
             bins = np.arange(0.25,6.75,0.5),
             color="#22bb99",
             ax=ax1)
ax1.set_title("Historgramm",fontsize=14,pad=15)
ax1.set_xlabel("Endnoten",fontsize=12,labelpad=15)
ax1.set_ylabel("Absolute Häufigkeit",fontsize=12,labelpad=10)
ax1.text(0.5,60,r"$\bar{x}=$"+"{m:0.2f}".format(m=mittelwert),fontsize=12)
ax1.text(0.5,55,r"$s=$"+"{s:0.2f}".format(s=std),fontsize=12)

ax2 = fig.add_subplot(1,2,2)
sns.boxplot(data=punkteNew,
            x="Note gerundet",
            width = 0.3,
            color="#22bb99",
            ax=ax2)
ax2.set_xlim(-0.2,6.4)
ax2.text(0.3,-0.37,r"$x_{50}=$"+"{m:0.2f}".format(m=x_50),fontsize=12)
ax2.text(0.3,-0.28,r"$IQR=$"+"{s:0.2f}".format(s=IQR),fontsize=12)
ax2.set_title("Boxplot",fontsize=14,pad=15)
ax2.set_xlabel("Endnoten",fontsize=12,labelpad=15)

plt.show()
```

***Bemerkung:*** Aus den Box-Plots ist ersichtlich, dass 75% der Studierenden bei den Aufgaben 6 und 9 weniger oder gleich 2 Punkte erreicht haben. Das hat gezeigt, dass in der Prüfung eine systematische Schwierigkeit vorhanden war. Die Notenskala wurde dann entsprechende angepasst.
